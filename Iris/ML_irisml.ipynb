```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp=MLPClassifier(random_state=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
trainX, testX, trainY, testY = train_test_split(data, target, test_size=0.3)
```

```python
trainY
```

```python
testY
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(data)
```

```python
trainX_std=scaler.transform(trainX)
```

```python
testX_std=scaler.transform(testX)
```

```python
mlp.fit(trainX_std, trainY)
```

```python
pred=mlp.predict(testX_std)
```

```python
print((testY!=pred).sum())
```

```python
from sklearn.metrics import accuracy_score
```
